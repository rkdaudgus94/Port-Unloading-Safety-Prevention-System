```python
img_list = glob('C:/safe/lie/*.jpg')
print(len(img_list))

txt_list = glob('C:/safe/lie/*.txt')
print(len(txt_list))

img_names = [os.path.basename(img_file) for img_file in img_list]
txt_names = [os.path.basename(txt_file) for txt_file in txt_list]

print("이미지 파일 이름:")
for img_name in img_names:
    print(img_name)

print("\n텍스트 파일 이름:")
for txt_name in txt_names:
    print(txt_name)
```

```python
from IPython.core.magic import register_line_cell_magic

# register_line_cell_mageic을 이용해 IPython에서 바로 사용할 수 있게 정의
@register_line_cell_magic
def writetemplate(line, cell):
  with open(line, 'w') as f:
    f.write(cell.format(**globals()))
```
